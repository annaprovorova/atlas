nilearn is a package that makes it easy to use advanced machine learning techniques to analyze data acquired with MRI machines.

**nilearn comes in with some data that are commonly used in neuroimaging. For instance, it comes with volumic template images of brains such as MNI:**

```python
import nilearn
from nilearn import datasets #     ,   plt
from nilearn import plotting
```

```python
print(nilearn.datasets.MNI152_FILE_PATH)
```

```python
nilearn.plotting.plot_img(nilearn.datasets.MNI152_FILE_PATH)
```

```python
nilearn.datasets.load_mni152_template(resolution=None)
```

```python
from nilearn.datasets import load_mni152_template
from nilearn.plotting import plot_img
img = load_mni152_template()
# display is an instance of the ZSlicer class
display = plot_img(img, display_mode='z')
```

```python
import numpy as np
#     ndarray
cut_coord = np.arange(-30, 70, 4)
display = plot_img(img, display_mode='z', cut_coords=cut_coord)
print(display.brain_color)
display.savefig('res.png')
```
