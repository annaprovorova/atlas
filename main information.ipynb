Coordinate system introduced by Talairach and Tournoux. The anterior commissure (AC) and posterior commissure (PC) are the landmarks used for developing this standard reference coordinate system. Its origin was defined at the AC, with x-and y-axis defining the horizontal plane and z-axis defining the vertical axis. The original figure [15] is slightly modified and reproduced with permission.

(Система координат, введенная Талайрахом и Турну. Передняя спайка(комиссура) (ПC) и задняя спайка (ЗC) являются ориентирами, используемыми для разработки этой стандартной эталонной системы координат. Его начало было определено в ПC, с осью x и y, определяющей горизонтальную плоскость, и осью z, определяющей вертикальную ось)

Передняя комиссура - пучок нервных волокон конечного мозга, соединяющий некоторые отделы обонятельного мозга (гиппокампальные извилины) с крючками височных долей. Передняя комиссура находится под клювом мозолистого тела позади концевой пластинки.

nilearn is a package that makes it easy to use advanced machine learning techniques to analyze data acquired with MRI machines.

**nilearn comes in with some data that are commonly used in neuroimaging. For instance, it comes with volumic template images of brains such as MNI:**

```python
import nilearn
from nilearn import datasets #важно сделать отдельно эти два импорта, как с plt
from nilearn import plotting
```

```python
print(nilearn.datasets.MNI152_FILE_PATH)
```

```python
nilearn.plotting.plot_img(nilearn.datasets.MNI152_FILE_PATH)
```

```python
nilearn.datasets.load_mni152_template(resolution=None)
```

```python
from nilearn.datasets import load_mni152_template
from nilearn.plotting import plot_img
img = load_mni152_template()
# display is an instance of the ZSlicer class
display = plot_img(img, display_mode='z')
```

```python
import numpy as np
#зададим сами координаты срезов через ndarray
cut_coord = np.arange(-30, 70, 4)
display = plot_img(img, display_mode='z', cut_coords=cut_coord)
print(display.brain_color)
display.savefig('res.png')
```

```python
display = plot_img(img, display_mode='z', cut_coords=[0])
```
